```python
import pandas as pd
import numpy as np
```

#### Preprocessing

```python
data = pd.read_csv("../../data/genedata.csv")
data
```

```python
data['class'] = data['class'].apply(lambda x: x-1)
data
```

```python
preprocessed_data = data.drop(columns=['id', 'class'])
preprocessed_data
```

## k-means clustering

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples, silhouette_score, calinski_harabasz_score, davies_bouldin_score, normalized_mutual_info_score
import matplotlib.pyplot as plt
import matplotlib.cm as cm
```

```python
range_n_clusters = [2,3,4,5,6,10]
```

### Clustering execution

k-means algorithm is applied storing the corresponding metrics.

```python
total_labels = [] # Labels computed by the clustering are going to be stored here.

silhouette_values = []
silhouette_avgs = []
ch_metrics = []
db_metrics = []
nmi_metrics = []

for n_cluster in range_n_clusters:
    km = KMeans(n_cluster, random_state=123)
    labels = km.fit_predict(preprocessed_data)
    total_labels.append(labels)
    
    # Silhouette analysis
    silhouette_avg = silhouette_score(data, labels)
    silhouette_avgs.append(silhouette_avg)

    silhouette_sample = silhouette_samples(data, labels)
    silhouette_values.append(silhouette_sample)
    
    # Calinski Harabasz analysis
    ch_metrics.append(calinski_harabasz_score(data, labels))

    # davies bouldin analysis
    db_metrics.append(davies_bouldin_score(data, labels))
    
    # Normalized mutual info analysis
    nmi_metrics.append(normalized_mutual_info_score(list(data['class']), labels, average_method="geometric"))
```

### Metrics

This code logs the results obtained in the indices after the clustering

Complementary function for logging the results

```python
def plot_cumulative_silhouette(data, labels, silhouette_values, n_clusters, silhouette_avg=None):
    fig, ax = plt.subplots()
    
    # Set limits for printing the silhouette.
    ax.set_xlim(-0.2, 0.4)
    ylim = len(data) + 10 * (n_clusters + 1)
    ax.set_ylim(0, ylim)

    y_lower = 10
    for i in range(n_clusters):
        # Compute the ith silhouette value and sort for each value.
        ith_cluster_silhouette_values = silhouette_values[labels == i]
        ith_cluster_silhouette_values.sort()

        size_cluster_i = ith_cluster_silhouette_values.shape[0]
        y_upper = y_lower + size_cluster_i

        color = cm.nipy_spectral(float(i) / n_clusters)
        ax.fill_betweenx(np.arange(y_lower, y_upper),
                         0, ith_cluster_silhouette_values,
                         facecolor=color, edgecolor=color, alpha=0.7)

        # Label the silhouette plots with their cluster numbers at the middle
        ax.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))

        # Compute the new y_lower for next plot
        y_lower = y_upper + 10

    # Labels for axes
    ax.set_xlabel("The silhouette coefficient values")
    ax.set_ylabel("Cluster label")

    # The vertical line for average silhouette score of all the values
    if silhouette_avg:
        ax.axvline(x=silhouette_avg, color="red", linestyle="--")
        ax.text(silhouette_avg - 0.03, ylim + 5, str(np.round(silhouette_avg, 2)))

    # Ticks for x and y axes
    ax.set_yticks([])  # Clear the yaxis labels / ticks
    ax.set_xticks([-0.2, -0.1, 0, 0.2, 0.4, 0.6, 0.8, 1])

    plt.suptitle(("Silhouette analysis for K - Means clustering "
                  "with k = %d" % n_clusters), fontweight='bold')
```

```python
def plot_metric(metric_values, range_of_clusters, metric_name, y_axe_label, print_avg=True):
    fig, ax = plt.subplots()

    plt.title(metric_name)
    ax.set_xlabel("Number of clusters")
    ax.set_ylabel(y_axe_label)

    ax.plot(range_of_clusters, metric_values, '-ok')
    
    if print_avg:
        avg_metric = np.mean(metric_values)
        ax.hlines(avg_metric, xmin=range_of_clusters[0], xmax=range_of_clusters[-1], color='r')
        ax.text(x=range_of_clusters[-1] + 0.5, y=avg_metric - 0.001, s=str(np.round(avg_metric, 3)))
```

```python
for i, n_cluster in enumerate(range_n_clusters):
    print(f"---- K-Means with k = {n_cluster} -----")
    print(f"* Silhouette avg -> {np.round(silhouette_avgs[i], 5)}")
    print(f"* Calinski-Harabasz -> {np.round(ch_metrics[i], 5)}")
    print(f"* Davies-Bouldin -> {np.round(db_metrics[i], 5)}")
    print(f"* NMI -> {np.round(nmi_metrics[i], 5)}")
    print()
    plot_cumulative_silhouette(preprocessed_data, total_labels[i], silhouette_values[i], n_cluster, silhouette_avgs[i])
```

In this part the different metrics are ploted, for the different number of clusters.

```python
plot_metric(silhouette_avgs, range_n_clusters, "Silhouette score", "Silhouette score")
plot_metric(ch_metrics, range_n_clusters, "Calinski-Harabasz index", "Calinski-Harabasz index")
plot_metric(db_metrics, range_n_clusters, "Davies-Bouldin index", "Davies-Bouldin index")
plot_metric(nmi_metrics, range_n_clusters, "Normalized Mutual Information", "Normalized Mutual Information score")
```

### Normalized mutual information

Calculate performances of mutual information

```python
def get_points(nmi):
    if (nmi >= 0.98):
        return 10
    elif (nmi >= 0.96):
        return 9
    elif (nmi >= 0.94):
        return 8
    elif (nmi >= 0.92):
        return 7
    elif (nmi >= 0.9):
        return 6
    elif (nmi >= 0.88):
        return 5
    elif (nmi >= 0.86):
        return 4
    elif (nmi >= 0.84):
        return 3
    elif(nmi >= 0.82):
        return 2
    elif (nmi >= 0.8):
        return 1
    else:
        return 0
        
```

```python
nmi_points = [(nmi, get_points(nmi), i) for i, nmi in enumerate(nmi_metrics)]
dtype = [('nmi', float), ('points', float), ('index', int)]

nmi_points = np.array(nmi_points, dtype=dtype)
sorted_executions = np.sort(nmi_points, order='nmi')[::-1]

for execution in sorted_executions:
    print(f"k = {range_n_clusters[execution[2]]} | nmi {np.round(execution[0], 5)} | points {execution[1]}")
```

### Agglomerative herarchical clustering

```python
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster
```

```python
linkage_metrics = ['single', 'complete', 'average', 'centroid', 'median', 'ward']
range_n_clusters = range(2, 11)

silhouette_avgs = []
nmi_metrics = []

for linkage_metric in linkage_metrics:
    X = linkage(preprocessed_data.to_numpy(), linkage_metric)
    log_results = True
    
    nmi_per_metric = []
    
    for n_clusters in range_n_clusters:
        labels = fcluster(X, n_clusters, criterion="maxclust")

        # NMI analysis
        nmi_per_metric.append(normalized_mutual_info_score(list(data['class']), labels, average_method="geometric"))
        
    nmi_metrics.append(nmi_per_metric)
```

The execution of the NMI is printed

```python
nmi_points = []

for metric_index, nmis in enumerate(nmi_metrics):
    for pos_index, nmi in enumerate(nmis):
        nmi_points.append((nmi, get_points(nmi), pos_index, metric_index))

dtype = [('nmi', float), ('points', float), ('pos_index', int), ('metric_index', int)]

nmi_points = np.array(nmi_points, dtype=dtype)
sorted_executions = np.sort(nmi_points, order='nmi')[::-1]

for execution in sorted_executions:
    print(f"k = {range_n_clusters[execution[2]]} | metric {linkage_metrics[execution[3]]} | nmi {np.round(execution[0], 5)} | points {execution[1]}")
```
